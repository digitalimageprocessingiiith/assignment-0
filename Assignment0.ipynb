```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
img1=cv2.imread("/content/drive/MyDrive/src/Jaguar.jpg")
img2=cv2.imread("/content/drive/MyDrive/src/Jungle.jpg")
img3=cv2.imread("/content/drive/MyDrive/src/mountain.jpg")
img4=cv2.imread("/content/drive/MyDrive/src/flower.webp")
images=[img1,img2,img3,img4]
```

```python
f, axarr = plt.subplots(2,2)
axarr[0,0].axis('off')
axarr[0,1].axis('off')
axarr[1,0].axis('off')
axarr[1,1].axis('off')
axarr[0,0].imshow(images[0])
axarr[0,1].imshow(images[1])
axarr[1,0].imshow(images[2])
axarr[1,1].imshow(images[3])

```

```python
f, axarr2 = plt.subplots(2,2)
axarr2[0,0].axis('off')
axarr2[0,1].axis('off')
axarr2[1,0].axis('off')
axarr2[1,1].axis('off')
axarr2[0,0].imshow(cv2.cvtColor(images[0], cv2.COLOR_BGR2RGB))
axarr2[0,1].imshow(cv2.cvtColor(images[1], cv2.COLOR_BGR2RGB))
axarr2[1,0].imshow(cv2.cvtColor(images[2], cv2.COLOR_BGR2RGB))
axarr2[1,1].imshow(cv2.cvtColor(images[3], cv2.COLOR_BGR2RGB))
```
